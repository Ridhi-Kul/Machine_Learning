```python
import mysql.connector
```

```python
conn = mysql.connector.connect(host = 'localhost', user ='root', password = "", database = "world")
# Parameters: 1. where is the database running (localhost, web server ID)
#             2. username on the machine
#             3. password
#             4. database name
```

The above commands create a mySQL connection object which is used to work further.

```python
df = pd.read_sql_query("SELECT * FROM city WHERE CountryCode LIKE 'USA'", conn) # Pass the sql query and the connection object to internally create a pandas DataFrame
```

### A lot of the parameters that can be passed on to read_sql_query() are similar as read_csv()

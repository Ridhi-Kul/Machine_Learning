# To load data directly from Github Server
### Click raw to get the url

```python
import requests
from io import StringIO

url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day15%20-%20working%20with%20csv%20files/aug_train.csv'
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/201001 Firefox/66.0"}
req = requests.get(url, headers = headers)
data = StringIO(req.text)

pd.read_csv(data)
```

```python
# OR YOU CAN ALSO DIRECTLY DO
pd.read_csv(url)
```

# Sep parameter

```python
# By default read csv takes seperator as ','. So tsv files can face error
url2 = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day15%20-%20working%20with%20csv%20files/movie_titles_metadata.tsv"
pd.read_csv(url2, sep = '\t', names = ['sno', 'name', 'release_year', 'rating', 'votes', 'genres'])

# We have used parameter names since it was taking the first row as the column name for DataFrame
```

# Index_col parameter

```python
pd.read_csv(url, index_col = 'enrollee_id')
```

# use_cols parameter
Help us import only required columns


```python
pd.read_csv(url, usecols = ['enrollee_id', 'gender', 'education_level'])
```

# Squeeze parameter
If we want only 1 parameter then instead of a pandas DataFrame we make it a pandas Series

I think it has been discontinued or its syntax has changed in recent python documentation since we are getting error and are unable to find it in official documentation as well.

```python
pd.read_csv(url, usecols = ['gender'], squeeze = True)
```

Instead to achieve same result we now use a method of the DataFrame object called .squeeze() to do so

```python
pd.read_csv(url, usecols = ['gender']).squeeze()
```

# Skiprows / nrows parameter
Instead of directly passing row we can also pass a lambda function in skiprows.

```python
pd.read_csv(url, skiprows = [0,5])
```

nrows only bring the number of rows that are mentioned.

Used especially in case of very very large dataset where the entire data points can't load in the RAM in one go.

```python
pd.read_csv(url, nrows = 100)
```

# Encoding parameter
If the dataset is not following utf-8 encoding then we can get a Unicode Decode Error. So in that case we need to use

pd.read_csv(filename, encoding = encoding_format_of_file)

* UTF-8 is a way to convert characters into bytes so computers can store and transmit text. It uses 1 to 4 bytes per character.

Can also do in a code editor.

# Handling Dates
Usually they come as string format preventing us from using months, dates, etc information

```python
url4 = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day15%20-%20working%20with%20csv%20files/IPL%20Matches%202008-2020.csv'
```

```python
pd.read_csv(url4)
```

```python
pd.read_csv(url4).info()
```

```python
pd.read_csv(url4, parse_dates = ['date']).info()
```

If we have separate columns that we want as date format then we can pass as [1,2,3].

But if we have separate columns for date, month then we can also combine it as [[1,3]] (list of list) or as {'new_date': [1,3]} (list in dictionary).

# Convertors
Say we want short names instead of full team name

```python
pd.read_csv(url4)
```

```python
def rename(name):
  if name == "Royal Challengers Banglore":
    return "RCB"
  else:
    return name
```

```python
rename("Royal Challengers Banglore")
```

```python
pd.read_csv(url4,converters={'team1':rename}) # Name of column on which transformation needs to be applied
```

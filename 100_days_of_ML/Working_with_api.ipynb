Now we see that 1 page has 20 films. We have 428 number of pages so we will use a loop and hit the api page number of times to fetch all the films on every page.


Can get the number of pages by using a json data viewer like an Online JSON viewer.

```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range (1, 429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=<<api_key>>&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
  df = df.append(temp_df, ignore_index = True) # We are ignoring index as page 1 has index 0-19 then so does page 2 from 0-19 and it keeps on going making the final df weird
```

### Newer pandas version do NOT support append() function. Instead

### * Method 1: use concat which works with pandas (pd.concat()) and not df. Pass 2 dataframe in the parameters since we need 2 objects to concat.

```python
for i in range (1, 429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=<<api_key>>&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
  df = pd.concat([df, temp_df], ignore_index=True) # We are ignoring index as page 1 has index 0-19 then so does page 2 from 0-19 and it keeps on going making the final df weird
```

```python
df
```

## However we have quicker best practice as

```python
dfs = []

for i in range(1, 429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=<<api_key>>&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)
```

# 7. How is the correlation between cols?
#### This tells us how the increase and decrease in 1 thing impacts another

#### Generally we try to remove the columns that don't impact the target. It is applied on NUMERICAL columns.

#### This step is performed multiple times during the entire process

```python
numeric_cols = ['PassengerId','Survived',	'Pclass',	'Age', 'SibSp', 'Parch',	'Fare']
df_numeric = df[numeric_cols]
df_numeric.corr()
```

### Here the value ranges from -1 to 1. Negavtive symbol represents inverse relationship and positive shows direct correlation.

```python
df_numeric.corr()['Survived']
```

Very small corr is in a way insignificant.
